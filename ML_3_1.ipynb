```python
from sklearn.model_selection import train_test_split


X_train, X_test, y_train, y_test = train_test_split(X, y,
                                                    test_size=0.33,
                                                    shuffle=True,
                                                    random_state=42)
```

Изучите документацию sklearn, посвященную классу LogisticRegression. Какую еще информацию можно вывести для обученной модели? Попробуйте изменить аргументы при создании модели и посмотрите, как это влияет на качество предсказания.

```python
from sklearn.metrics import accuracy_score

model_1 = LogisticRegression(solver='liblinear')
model_1.fit(X_train, y_train)
preds_1 = model_1.predict(X_test)
print("Original Model Accuracy:", accuracy_score(y_test, preds_1))

model_2 = LogisticRegression(penalty='l1', solver='liblinear')
model_2.fit(X_train, y_train)
preds_2 = model_2.predict(X_test)
print("L1 Regularization Model Accuracy:", accuracy_score(y_test, preds_2))

model_3 = LogisticRegression(C=0.01, solver='liblinear')
model_3.fit(X_train, y_train)
preds_3 = model_3.predict(X_test)
print("High Regularization Model Accuracy:", accuracy_score(y_test, preds_3))
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, confusion_matrix
from sklearn import svm, neighbors, neural_network, tree, naive_bayes, linear_model, ensemble
import matplotlib.pyplot as plt
import seaborn as sns

models = {
    "SVM (No Kernel)": svm.LinearSVC(),
    "SVM (Linear)": svm.SVC(kernel='linear'),
    "SVM (Poly)": svm.SVC(kernel='poly'),
    "SVM (RBF)": svm.SVC(kernel='rbf'),
    "KNN": neighbors.KNeighborsClassifier(),
    "MLP": neural_network.MLPClassifier(),
    "Decision Tree": tree.DecisionTreeClassifier(),
    "Naive Bayes": naive_bayes.GaussianNB(),
    "Passive-Aggressive": linear_model.PassiveAggressiveClassifier(),
    "Ridge": linear_model.RidgeClassifier(),
    "Random Forest": ensemble.RandomForestClassifier(),
    "Bagging": ensemble.BaggingClassifier()
}

fig, axes = plt.subplots(nrows=len(models), figsize=(10, 5 * len(models)))

for ax, (name, model) in zip(axes, models.items()):
    model.fit(X_train, y_train)
    predictions = model.predict(X_test)
    cm = confusion_matrix(y_test, predictions)

    sns.heatmap(cm, annot=True, fmt='g', ax=ax, cmap='Blues')
    ax.set_title(name)
    ax.set_xlabel('Predicted labels')
    ax.set_ylabel('True labels')

plt.tight_layout()
plt.show()
```

```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split

data = load_iris()
X = data.data
y = data.target

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42, shuffle=True)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix
from sklearn import svm, neighbors, neural_network, tree, naive_bayes, linear_model, ensemble
import seaborn as sns
import matplotlib.pyplot as plt

models = {
    "SVM (No Kernel)": svm.LinearSVC(),
    "SVM (Linear)": svm.SVC(kernel='linear'),
    "SVM (Poly)": svm.SVC(kernel='poly'),
    "SVM (RBF)": svm.SVC(kernel='rbf'),
    "KNN": neighbors.KNeighborsClassifier(),
    "MLP": neural_network.MLPClassifier(),
    "Decision Tree": tree.DecisionTreeClassifier(),
    "Naive Bayes": naive_bayes.GaussianNB(),
    "Passive-Aggressive": linear_model.PassiveAggressiveClassifier(),
    "Ridge": linear_model.RidgeClassifier(),
    "Random Forest": ensemble.RandomForestClassifier(),
    "Bagging": ensemble.BaggingClassifier()
}

fig, axes = plt.subplots(nrows=len(models), figsize=(10, 5 * len(models)))

for (name, model), ax in zip(models.items(), axes):
    model.fit(X_train, y_train)
    predictions = model.predict(X_test)

    acc = accuracy_score(y_test, predictions)
    cm = confusion_matrix(y_test, predictions)

    print(name)
    print("-" * len(name))
    print("Accuracy:", acc)
    sns.heatmap(cm, annot=True, ax=ax, cmap='Blues', fmt='g')
    ax.set_title(name)
    ax.set_xlabel('Predicted labels')
    ax.set_ylabel('True labels')
    print("\n")

plt.tight_layout()
plt.show()
```
